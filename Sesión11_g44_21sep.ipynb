```python
## Librerías Básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Librerías Especializadas
from statsmodels.tsa.seasonal import seasonal_decompose
## Tamaño gráfica (Largo, alto)
plt.rcParams["figure.figsize"] = (10,6)
```

```python
## Datos
pasajeros = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/airline_passengers.csv',index_col='Month', parse_dates=True)
```

```python
pasajeros.head()
```

```python
# Renombrar 
pasajeros = pasajeros.rename(columns={'Thousands of Passengers':'Miles de pasajeros'})
```

```python
pasajeros.head()
```

```python
## Graficar con pandas
pasajeros['Miles de pasajeros'].plot()
plt.xlabel('Mes')
plt.ylabel('Miles')
plt.title('Pasajeros mensuales E.U. 1949-1960');
```

La serie tiene tendencia creciente, tiene componente estacional ya que se observa que los pasajeros siempre se incrementan en los meses de julio-agosto

## Modelo ETS (Error-Trend-Seasonally)
Modelo Irregular-Tendencia(ciclo)-Estacional

```python
ets = seasonal_decompose(pasajeros['Miles de pasajeros'], model='mul')
```

```python
ets.plot()
plt.show()
```

```python
ets.trend.plot();
```

```python
ets.seasonal.plot();
```

```python
ets.resid.plot();
```

```python
## Desestacionalizar la serie: quitarle el componente estacional
pasajeros['Desestacionalizada'] = pasajeros['Miles de pasajeros'] / ets.seasonal
```

```python
pasajeros['Desestacionalizada'].plot();
```
