```python
## Librerías de cajón
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## datos a trabajar 
edades = [23,25,30,19]
personas = ['Juan','Pedro','Ana','Lissa']

dicc_1 = {'municipio':['Escobedo','Apodaca','Monterrey','Santiago'],'nacimientos':[200,350,400,120],'muertes':[20,35,40,18],'matrimonios':[56,68,70,19]}
datos_mun = pd.DataFrame(dicc_1)
```

```python
datos_mun
```

```python
edades, personas
```

```python
## gráfica semana pasada (Sesión4)
plt.bar(personas, edades)
plt.title('Edades de 4 personas')
plt.xlabel('Nombre')
plt.ylabel('Edad en años')
plt.show()
```

```python
plt.bar(datos_mun['municipio'], datos_mun['matrimonios'])
plt.title('Matrimonios en N.L.')
plt.xlabel('Municipio')
plt.ylabel('Cantidad matrimonios')
plt.show()
```

```python
# Dos gráficas juntas 
fig,(graf1, graf2) = plt.subplots(2,1)
```

```python
# Dos gráficas juntas 
fig,(graf1, graf2) = plt.subplots(1,2)
```

```python
# Dos gráficas juntas 
fig,(graf1, graf2) = plt.subplots(1,2)

# Gráfica 1 edades
graf1.bar(personas,edades)

# Gráfica 2 matrimonios
graf2.bar(datos_mun['municipio'],datos_mun['matrimonios'])
plt.show()
```
