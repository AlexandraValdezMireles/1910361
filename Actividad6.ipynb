```python
## Librerías Básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Librerías Especializadas
from lifelines import CoxPHFitter, KaplanMeierFitter, ExponentialFitter, WeibullFitter, LogNormalFitter
import seaborn as sns
```

```python
## Importar Datos 
df_desercion = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/desercion_laboral.csv')
```

```python
df_desercion.head()
```

### DEPURAR LA BASE DE DATOS Y ESTADISTICAS DESCRIPTIVAS

```python
## Verificar missing values (valores perdidos)
df_desercion.isnull().sum()
```

```python
## Verificar que no haya valores duplicados
df_desercion.duplicated().sum()
```

```python
## Eliminar valores duplicados
df_desercion=df_desercion.drop_duplicates(keep='first').reset_index(drop=True)
```

```python
df_desercion.head()
```

```python
## Verificar nuevamente que no haya valores duplicados
df_desercion.duplicated().sum()
```

```python
## Verificar que las variables sean numéricas
df_desercion.dtypes
```

```python
## Convertir a dummies
dummies=pd.get_dummies(df_desercion['departamento'], prefix='depa').astype(int)
dummies.head()
```

```python
dummies.describe()
```

```python
#Se guardan los dummies en el dataframe y se eliminan los datos reales de departamento
dummies = dummies[['depa_Investigación','depa_administración','depa_contabilidad', 'depa_desarrollo de producto', 'depa_informática', 'depa_mantenimiento', 'depa_mercadotecnia', 'depa_recursos humanos', 'depa_soporte']]
df_desercion = pd.concat([df_desercion,dummies],axis=1)
df_desercion = df_desercion.drop('departamento',axis=1)
```

```python
df_desercion.head()
```

```python
## Convertir a dummies
dummies=pd.get_dummies(df_desercion['salario'], prefix='sal').astype(int)
dummies.head()
```

```python
dummies.describe()
```

```python
#Se guardan los dummies en el dataframe y se eliminan los datos reales de salario
dummies = dummies[['sal_bajo', 'sal_medio']]
df_desercion = pd.concat([df_desercion,dummies],axis=1)
df_desercion = df_desercion.drop('salario',axis=1)
```

```python
df_desercion.head()
```

```python
## Estadisticas Descriptivas
df_desercion.describe()
```

### ESTIMAR MODELO COX

```python
cph = CoxPHFitter().fit(df_desercion, 'semestre','renunciar')
cph.print_summary()
```

```python
cph.hazard_ratios_
```

**satisfaccion:** Su pvalue es menor que alfa, rechazamos Ho significa que si el empleado está satisfecho con su trabajo disminuye el riesgo de renuncia, por cada escala de satisfacción el riesgo de renuncia disminuye en un 78%

**evaluación:** Su pvalue es mayor que alfa, no rechazamos Ho significa que si el empleado realiza bien su trabajo no tiene efecto en el riesgo de renuncia. 

**proyectos:** Su pvalue es menor que alfa, rechazamos Ho significa que el número de proyectos asignados al empleado disminuye el riesgo de renuncia, por cada proyecto asignado disminuye el riesgo de renuncia en un 75%. 

**promedio_horas_mes:** Su pvalue es menor que alfa, rechazamos Ho significa que el promedio de horas mensuales trabajadas por el empleado aumenta el riesgo de renuncia en un 0.18%. 

**accidente:** Su pvalue es menor que alfa, rechazamos Ho significa que si el empleado ha tenido un accidente dentro de la empresa disminuye el riesgo de renuncia en un 29% comparado a si no ha tenido un accidente. 

**ascenso_5years:** Su pvalue es menor que alfa, rechazamos Ho significa que si el empleado ha sido ascendido durante los últimos 5 años disminuye el riesgo de renuncia en un 25% comparado a si no ha sido ascendido durante los últimos 5 años. 

**depa_Investigación:** Su pvalue es menor que alfa, rechazamos Ho significa que si el empleado trabaja en el departamento de investigación disminuye el riesgo de renuncia en un 68% comparado a trabajar en el departamento de ventas. 

**depa_administración, depa_contabilidad, depa_desarrollo de producto, depa_informática, depa_mantenimiento, depa_mercadotecnia, depa_recursos humanos, depa_soporte:** Su pvalue es mayor que alfa, no rechazamos Ho significa que si el empleado trabaja en cualquiera de estos departamentos no tiene efecto sobre el riesgo de renuncia. 

**sal_bajo:** Su pvalue es menor que alfa, rechazamos Ho significa que si el salario del empleado es bajo aumenta el riesgo de renuncia en un 359% comparado a si su salario es alto. 

**sal_medio:** Su pvalue es menor que alfa, rechazamos Ho significa que si el salario del empleado es medio aumenta el riesgo de renuncia en un 213% comparado a si su salario es alto. 

#### **RECOMENDACIÓN A LA EMPRESA:** 
Que traten de mantener a sus empleados de los distintos departamentos ocupados asignandoles diversos proyectos para que así ellos se sientan útiles en el trabajo y al igual se sientan satisfechos con ello, para que a su vez puedan recibir un buen salario y lograr ser ascendidos en un corto periodo.

```python
## Estimar el modelo AFT
## Primero tenemos que determinar si es exponencial, weibull o lognormal
```

```python
from lifelines import ExponentialFitter, WeibullFitter, LogNormalFitter
```

```python
mexpo = ExponentialFitter().fit(df_desercion['semestre'], df_desercion['renunciar'])
mweibull = WeibullFitter().fit(df_desercion['semestre'], df_desercion['renunciar'])
mlogn = LogNormalFitter().fit(df_desercion['semestre'], df_desercion['renunciar'])
```

```python
mexpo.AIC_.round(2), mweibull.AIC_.round(2), mlogn.AIC_.round(2)
```
